## Project 🚧

The marketing team needs help on a new project. They are experiencing a decrease in the number of matches, and they are trying to find a way to understand **what makes people interested into each other**. 

They decided to run a speed dating experiment with people who had to give Tinder lots of informations about themselves that could ultimately reflect on ther dating profile on the app.

Tinder then gathered the data from this experiment. Each row in the dataset represents one speed date between two people, and indicates wether each of them secretly agreed to go on a second date with the other person.

## Scope of this project 🖼️

Data was gathered from participants in experimental speed dating events from 2002-2004. During the events, the attendees would have a four minute "first date" with every other participant of the opposite sex. At the end of their four minutes, participants were asked if they would like to see their date again. They were also asked to rate their date on six attributes: Attractiveness, Sincerity, Intelligence, Fun, Ambition, and Shared Interests.

The dataset also includes questionnaire data gathered from participants at different points in the process. These fields include: demographics, dating habits, self-perception across key attributes, beliefs on what others find valuable in a mate, and lifestyle information.
```python
print(df.info())
display(df.describe(include='all'))
#print(df.describe(include='all'))
print(f"nombre de cellules manquantes : {df.isnull().sum().sum()}")
print(f"le nombre de lignes avec des cellules manquantes : {df.isnull().any(axis=1).sum()}")
print(df['gender'].value_counts())
```

```python
nb_participants  = len(df['iid'].unique())
nb_sessions = len(df['wave'].unique())
nb_rdv = (len(df))/2  # chaque rendez-vous est enregistré deux fois (une fois par participant)
print(f"Nombre de sessions :{nb_sessions}, {nb_participants} participants, pour un total de {nb_rdv} rendez-vous.")
moy_participants_per_session = round(nb_participants / nb_sessions, 2)
print(f"Nombre moyen de participants par session : {moy_participants_per_session}")
nb_matches = (df['match'].sum())/2
moy_matches_per_session = round(nb_matches / nb_sessions, 2)
print(f"Nombre total de 'matchs' : {nb_matches}, soit en moyenne par session: {moy_matches_per_session} ")
moy_rdv_per_session = round(nb_rdv / nb_sessions, 2)
print(f"Nombre moyen de rdv par session : {moy_rdv_per_session}, soit taux de conversion de  {round(moy_matches_per_session*100/moy_rdv_per_session, 2)}% par session.")

```

### Repartision par Age et genre de la population d'étude

```python
df_unique = df.drop_duplicates(subset=['iid'])
df_unique['gender_label'] = df_unique['gender'].map({0: 'Femme', 1: 'Homme'})
fig = px.histogram(df_unique, x='age', color='gender_label', barmode='group',
                   title='Age Distribution by Gender',
                   color_discrete_map={'Femme': '#EF553B',  # Rouge pour 0 (homme)
                                      'Homme': '#636EFA'},  # Bleu pour 1 (femme))
                    text_auto=True)  # Affiche les valeurs sur les barres
fig.show()
```

```python
# Créer un tableau de contingence
tableau = pd.crosstab(df_unique['age'], df_unique['gender_label'])
print(tableau)

# Pour voir une ligne spécifique (ex: 27 ans)
print(f"\nÀ 27 ans:")
print(tableau.loc[27])
```

### Comptage Homme / Femme par wave

```python
df_compt_HF = df_unique.groupby('wave')['gender'].value_counts().unstack()    
df_compt_HF
```

```python
fig = px.pie(df_unique, names='gender_label',
                   title='Age Distribution by Gender',
                   color_discrete_map={'Femme': '#636EFA',  # Rouge pour 0 (homme)
                                      'Homme': '#EF553B'})  # Bleu pour 1 (femme))
fig.update_traces(texttemplate='%{label}<br>%{value} (%{percent})')
fig.show()
```

### Verification participation unique (Verification qu'un participant n'a pas participé a plusieur session)


```python
# Compter combien de waves différentes pour chaque iid
waves_par_iid = df.groupby('iid')['wave'].nunique()
# Filtrer ceux qui sont dans plus d'une wave
iid_multiples = waves_par_iid[waves_par_iid > 1]
print(f" Nombre de participants dans plusieurs waves: {iid_multiples.count()}")


```

### importance de la race, de la religion, du satus social dans le choix
